# Working with Missing Data

Car crashes 2018

```python
import pandas as pd

mvc = pd.read_csv("nypd_mvc_2018.csv")
mvc
```

```python
# Count missing values
null_counts = mvc.isnull().sum()
null_counts
```

```python
# Can fix missing total killed values by adding values in pedestrians_killed, cyclist_killed and motorist_killed
# as those columns have no missing values
killed_cols = [col for col in mvc.columns if "killed" in col]
killed_df = mvc[killed_cols].copy()
killed_manual_sum = killed_df.iloc[:,:3].sum(axis=1)
killed_mask = mvc["total_killed"] != killed_manual_sum
killed_non_eq = killed_df[killed_mask]
killed_non_eq
```
