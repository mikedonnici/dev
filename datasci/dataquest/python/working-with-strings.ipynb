````python
import pandas as pd

happiness2015 = pd.read_csv("world-happiness-2015.csv")
world_dev = pd.read_csv("world_dev.csv")

happiness2015.head(3)
````

````python
world_dev.head(3)
````

````python
# Merge (horizontal join) dataframes 
merged = pd.merge(left=happiness2015, right=world_dev, how="left", left_on="Country", right_on="ShortName")
merged
````

````python
col_renaming = {'SourceOfMostRecentIncomeAndExpenditureData': 'IESurvey'}
merged = merged.rename(columns=col_renaming)
merged
````

````python
# Extract the last word of the currency description
def extract_last_word(element):
    return str(element).split()[-1]

merged["Currency Apply"] = merged["CurrencyUnit"].apply(extract_last_word)
merged["Currency Apply"]
````

````python
# As above, with vector method
merged["Currency Vectorized"] = merged["CurrencyUnit"].str.split().str.get(-1)
merged["Currency Vectorized"].head()
````

````python
# Find empty string values in CurrencyUnit
lengths = merged["CurrencyUnit"].str.len()
value_counts = lengths.value_counts(dropna=False)
value_counts
````

Above shows three benefits of using vectorised string methods:

1. Better performance
2. Code that is easier to write and read
3. Automatically excludes missing values

````python
# Matching strings with regex
# Find records containing the pattern [Nn]ational accounts in the notes
national_accounts = merged['SpecialNotes'].str.contains(r"[Nn]ational accounts")
national_accounts
````

````python
# Can't index using NaN values so make those False instead
filter = merged["SpecialNotes"].str.contains("[nN]ational accounts", na=False)
merged_national_accounts = merged[filter]
for i, row in merged_national_accounts.head().iterrows():
    print(f"{row['Country']}: {row['SpecialNotes']}")
````

````python
# Extract the string based on regex
# Extract a year value from the notes - extraction requires the use of a capturing group 
# which is a regex inside parentheses - r"(pattern)"
# Note that in (this) later version of pandas the expand param is set to True by default
# which will return a Dataframe. Here it is explicitly set to False to return a Series.
years = merged["SpecialNotes"].str.extract(r"([1-2][0-9]{3})", expand=False)
print(years.head(10))
print(type(years)) # is a series
````

````python
# expand=True (default) will retun a Dataframe
years_df = merged["SpecialNotes"].str.extract(r"([1-2][0-9]{3})", expand=True)
print(years_df.head(10))
print(type(years_df))
````

````python
# Using a named capturing group and extractall() to extract mor than one occurence
# First, set the Country as the index to make results easier to read
merged = merged.set_index("Country")
````

````python
merged.head()
````

````python
# regex with a named capturing group (?P<name>...)
pattern = r"(?P<Years>[1-2][0-9]{3})"
years = merged["IESurvey"].str.extractall(pattern)
years
# note the match column - it contains an index value (0,1,2..) for each match found
````

````python
value_counts = years["Years"].value_counts()
value_counts
````

````python
# Data has years in format yyyy and yyyy/yy, eg 2010/11
# Add capturing groups to include both formats
# Note there are no parentheses around /? so the resulting dataframe 
# will only contain First_Year and Second_Year colummns
pattern = r"(?P<First_Year>[1-2][0-9]{3})/?(?P<Second_Year>[0-9]{2})"
years = merged["IESurvey"].str.extractall(pattern)
years
````

````python
# Take the first two numbers from the first year and prepend to the second year
years["Second_Year"] = years["First_Year"].str[0:2] + years["Second_Year"]
years
````

````python
# Clean up and plot example...

# Remove leading and trailing whitespace
merged["IncomeGroup"] = merged["IncomeGroup"].str.strip()

# Replace some strings
replace = {
    "Upper middle income"  : "UPPER MIDDLE",
    "Lower middle income"  : "LOWER MIDDLE",
    "High income: OECD"    : "HIGH OECD",
    "Low income"           : "LOW",
    "High income: nonOECD" : "HIGH NONOECD"
}
for k, v in replace.items():
    merged["IncomeGroup"] = merged["IncomeGroup"].str.replace(k, v)

merged["IncomeGroup"]
````

````python
# Create a pivot table - displays mean() by default
pv_incomes = merged.pivot_table(index="IncomeGroup", values="Happiness Score")
pv_incomes
````

````python
# Plot
pv_incomes.plot(kind="bar", rot=30, ylim=(0,10))
````

````python
# Plot
pv_incomes.plot(kind="barh", xlim=(0,10))
````

### Working With Strings In Pandas


#### Regular Expressions

- Match multiple characters, specify the characters between "[ ]"
- This expression would match "national accounts" and "National accounts":
```python
pattern = r"[Nn]ational accounts"
```

- Match a range of characters
- This expression would match any number between 0 and 9:
```python
pattern = r"[0-9]"
```

- To create a capturing group, or indicate that only the character pattern matched should be extracted, specify the characters between "( )"
- This expression would match years.
```python
pattern = r"([1-2][0-9][0-9][0-9])"
```

- To repeat characters, use "{ }"
- This would match the pattern "[0-9]" three times, and also match years:
```python
pattern = r"([1-2][0-9]{3})"
```

- To name a capturing group, use `(?P<name>...)`
- This expression would match years and name the capturing group "Years":
```python    
pattern = r"(?P<Years>[1-2][0-9]{3})"
```

#### Vectorized String Methods

- Find specific strings or substrings in a column:
```python
df[col_name].str.contains(pattern)
```

- Extract specific strings or substrings in a column:
```python
df[col_name].str.extract(pattern)
```

- Extract more than one group of patterns from a column:
```python
df[col_name].str.extractall(pattern)
```

- Replace a regex or string in a column with another string:
```python
df[col_name].str.replace(pattern, replacement_string)
```

####  Concepts
- Pandas has built in a number of vectorized methods that perform the same operations for strings in Series as Python string methods.
- A regular expression is a sequence of characters that describes a search pattern. In pandas, regular expressions is integrated with vectorized string methods to make finding and extracting patterns of characters easier.
